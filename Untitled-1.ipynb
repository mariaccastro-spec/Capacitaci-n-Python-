```python
import os
os.chdirs.chdir(r'C:\Users\maria.c.castro\OneDrive - Accenture\Capacitación Python principiantes')

import pandas as pd
from datetime import datetime
 
# 1. Cargamos los archivos (Mezcla de CSV y Excel)
 
# El primer archivo es CSV
df1 = pd.read_csv('nomina_empleados.csv')
 
# Los otros son Excel (.xlsx).
# Por defecto lee la primera pestaña; sino usamos: sheet_name='Nombre'
df2 = pd.read_excel('nomina_ejercicio_limpieza_1.xlsx')
df3 = pd.read_excel('nomina_ejercicio_limpieza_2.xlsx')
df4 = pd.read_excel('nomina_ejercicio_limpieza_3.xlsx')
 
# 2. Elegimos nombres estándar para poder unir las columnas
df1 = df1.rename(columns={'ID_Emplea': 'ID', 'Sueldo_Bruto': 'Sueldo'})
df2 = df2.rename(columns={
    'ID_Emp': 'ID',
    'Nombre_Completo': 'Nombre',
    'Depto': 'Departamento',
    'Salario': 'Sueldo'
})
df3 = df3.rename(columns={
    'Codigo_Empleado': 'ID',
    'Empleado': 'Nombre',
    'Remuneracion': 'Sueldo'
})
df4 = df4.rename(columns={'Sueldo_Bruto': 'Sueldo'})
 
# 3. Unimos todo en un solo DataFrame
df = pd.concat([df1, df2, df3, df4], ignore_index=True)
 
print('Archivos integrados con éxito.')
print(f'Total de filas: {len(df)}')

# Ver todas las columnas sin que se corten
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)

# Ver las primeras filas como tabla
print(df.head())
```

```python
# 1. Primero limpiar strings
df['Nombre'] = df['Nombre'].str.strip().str.title()
df['Departamento'] = df['Departamento'].str.strip().str.upper()

# 2. Luego mapear (ahora sí todo está en mayúsculas y sin espacios extra)
mapeo_deptos = {
    'RRHH': 'RECURSOS HUMANOS',
    'IT': 'IT',          # 👈 sin espacio
    'CONTABILID': 'CONTABILIDAD',
    'RECURSOS HUMANOS': 'RECURSOS HUMANOS'  # por si ya viene bien
}
df['Departamento'] = df['Departamento'].replace(mapeo_deptos)
df['Departamento'] = df['Departamento'].str.strip().str.upper()

df['Departamento'].value_counts()  # 👈 para verificar que quedó limpio
```

```python
# 3. Limpiar la columna de Sueldo (si viene como texto con $)
if df['Sueldo'].dtype == 'object':
    df['Sueldo'] = df['Sueldo'].str.replace('$', '', regex=False) \
                               .str.replace(',', '', regex=False)
    df['Sueldo'] = pd.to_numeric(df['Sueldo'], errors='coerce')

df.head()
```
